```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("heart.csv")
df.head(15)
```

```python
# Kategorik özellikler
categorical_columns = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

# Veri çerçevesinin bir kopyasını al
df_encoded = df.copy()

# Label Encoding işlemi
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])  # Kategorik veriyi sayısal verilere dönüştür
    label_encoders[col] = le  # LabelEncoder'ı sakla

# X ve y oluşturuluyor
X = df_encoded.drop(columns=["HeartDisease"]).values.astype(float)  
y = df_encoded["HeartDisease"].values.astype(float)

# Veri çerçevesini bir CSV dosyasına kaydet
df_encoded.to_csv("encoded_data.csv", index=False)

print("Label encoding işlemi tamamlandı ve veri kaydedildi.")
```
